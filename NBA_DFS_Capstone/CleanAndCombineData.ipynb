```python
import pandas as pd
from glob import glob
import datetime
```

```python
games = pd.read_csv('AllBoxScores.csv', index_col= 0)
games = games[games['Starters'] != 'Team Totals']
games['date'] = games['date'].apply(lambda x: datetime.datetime.strptime(x,"%Y-%m-%d").strftime("%Y-%m-%d"))
games = games[((games['date'] <= '2017-04-12') | (games['date'] > '2017-05-01')) & (games['date'] <= '2018-04-11')]
#games = games.set_index(['Starters','date'])
games = games.drop('Starters.1',axis=1)
games = games.drop('MP.1',axis=1)
games
```

```python
dfs_frames  = [pd.read_csv(filename, index_col= 0) for filename in glob('*DFS.csv')]
dfs = dfs_frames[0].append(dfs_frames[1])

dfs_cols_to_keep = ['DateTime','DoubleDoubles','DraftKingsPosition','DraftKingsSalary','FantasyPointsDraftKings',
                    'GameID','HomeOrAway','InjuryBodyPart','InjuryNotes','InjuryStartDate','InjuryStatus',
                    'Name','Opponent','OpponentPositionRank','OpponentRank','Position','Season','Started',
                    'Team','TripleDoubles']
dfs = dfs[dfs_cols_to_keep]
```

```python
dates_and_times = dfs.DateTime.apply(str).apply(str.split)
dfs['Date'] = [dt[0] for dt in dates_and_times]
dfs['Time'] = [dt[1] for dt in dates_and_times]
dfs = dfs.drop('DateTime',axis=1)
```

```python
dfs_all = pd.merge(games,dfs,how='outer',left_on=['Starters','date'],right_on=['Name','Date'])
dfs_all['Player'] = dfs_all['Name'].fillna(dfs_all['Starters'])
dfs_all['Date'] = dfs_all['Date'].fillna(dfs_all['date'])
dfs_all = dfs_all.drop('Name',axis=1)
dfs_all = dfs_all.drop('Starters',axis=1)
dfs_all = dfs_all.drop('date',axis=1)
```

```python
pd.set_option('display.max_columns', 80)
dfs_all.loc[dfs_all['Player'] == 'CJ McCollum',:]
#dfs_all.loc[dfs_all['Player'] == 'C.J. McCollum',:].fillna(dfs_all.loc[dfs_all['Player'] == 'CJ McCollum',:],axis=1)
```

```python
dfs_all.loc[:,dfs_all.dtypes == object] = dfs_all.loc[:,dfs_all.dtypes == object].fillna('')
dfs_all.loc[:,dfs_all.dtypes != object] = dfs_all.loc[:,dfs_all.dtypes != object].fillna(0)
dfs_all
```

```python
cols = dfs_all.columns.tolist()
dfs_all = dfs_all[[cols[-1]] + cols[:len(cols)-1]]
dfs_all
```

```python
dfs_all.to_csv('CleanCombinedDFSAll.csv')
```
