```python
from bs4 import BeautifulSoup
import requests
import re

r = requests.get('https://www.basketball-reference.com/boxscores/201710170CLE.html')
soup = BeautifulSoup(r.text)

columns = [th.getText() for th in soup.findAll('tr')[1].findAll('th')]
adv_columns = [th.getText() for th in soup.findAll('tr')[17].findAll('th')]

months = ['october','november','december','january','february','march','april']
years = ['2017','2018']
url_bases = ['https://www.basketball-reference.com/leagues/NBA_' + year + '_games-' for year in years]
urls = [url_base + month + '.html' for month in months for url_base in url_bases]
links = []

for url in urls:
    r = requests.get(url)
    soup = BeautifulSoup(r.text,)
    month_links = [link.get('href') for link in soup.find_all('a')]
    links = links + [link for link in month_links if '/boxscores/2' in link]

links = ["https://www.basketball-reference.com" + link for link in links]
links
```

```python
games = pd.DataFrame()

for game in links:
    r = requests.get(game)
    soup = BeautifulSoup(r.text)
    data_rows = soup.findAll('tr')[2:]
    player_data = [[td.getText() for td in data_rows[i].findAll('td')] for i in range(len(data_rows))]
    player_data = [data for data in player_data if data != []]
    player_names = [[th.getText() for th in row.findAll('th')] for row in data_rows]
    player_names = [name for name in player_names if len(name) == 1]
    table_rows = [player_names[i] + player_data[i] for i in range(len(player_data))]
    basic_rows = [row for row in table_rows if (len(row) == len(columns))]
    basic_frame = pd.DataFrame(basic_rows, columns=columns)
    adv_rows = [row for row in table_rows if (len(row) == len(adv_columns))]
    adv_frame = pd.DataFrame(adv_rows, columns=adv_columns)
    game_box = pd.concat([basic_frame,adv_frame],axis=1)
    date = re.findall("\d+", game)[0][:8]
    game_box['date'] = datetime.datetime.strptime(date,"%Y%m%d").strftime("%Y-%m-%d")
    games = games.append(game_box)

```

```python
games.to_csv('AllBoxScores.csv')
```
