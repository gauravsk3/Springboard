```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

df = pd.read_csv('data/human_body_temperature.csv')
```

```python
# Your work here.
df.temperature.hist()
plt.show()
```

```python
sm.qqplot(df.temperature,line='s')
plt.show()
```

```python
def ecdf(data):
    n = len(data)
    x = np.sort(data)
    y = np.arange(1, n+1) / n
    return x, y

samples = np.random.normal(np.mean(df.temperature),np.std(df.temperature),size=10000)
x,y = ecdf(df.temperature)
plt.plot(x,y)
x_samp,y_samp = ecdf(samples)
plt.scatter(x_samp,y_samp,color='yellow')
```
