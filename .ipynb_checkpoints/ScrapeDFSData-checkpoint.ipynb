```python
import pandas as pd
```

```python
import http.client, urllib.request, urllib.parse, urllib.error, base64

headers = {
    # Request headers
    'Ocp-Apim-Subscription-Key': '9f2f1124447c42978f6ba81d55e5a6bf',
}

params = urllib.parse.urlencode({
})

def getDayStats(date):
    try:
        conn = http.client.HTTPSConnection('api.fantasydata.net')
        conn.request("GET", "/v3/nba/projections/JSON/PlayerGameProjectionStatsByDate/" + date + "?%s" % params, "{body}", headers)
        response = conn.getresponse()
        data = response.read()
        #print(data)
        conn.close()
        return data
    except Exception as e:
        print("[Errno {0}] {1}".format(e.errno, e.strerror))
```

```python
import datetime

start1617 = '25-OCT-2016'
end1617 = '12-APR-2017'
start1718 = '17-OCT-2017'
end1718 = '11-APR-2018'

def getDateRange(start, end):
    days = pd.date_range(start1718, end1718)
    days = pd.Series(days.format())
    days = days.apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d').strftime('%d-%b-%Y').upper())
    return days
```

```python
all_data = pd.DataFrame()
for day in days:
    all_data = all_data.append(pd.read_json(getDayStats(day)))

all_data
```
